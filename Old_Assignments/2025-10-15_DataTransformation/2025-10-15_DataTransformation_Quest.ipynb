```python
import pandas as pd
```

```python
student = pd.read_csv('student.csv')
student
```

```python
point = pd.read_csv('point.csv')
point
```

# 1. 데이터 결합과 변환을 연습해봅시다!

## 다음은 호그와트의 학생 명부와 점수 데이터입니다. 두 데이터를 합쳐서 데이터를 분석해봅시다.

```python
final = pd.merge(point, student, on='student_id', how='left').drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'])
```

```python
final.head()
```

```python
final.describe()
```

```python
final.info()
```

## 기숙사별 총점을 계산해서 어느 기숙사가 앞서고 있는지 확인합니다.

```python
final['house'].unique()
```

```python
final.groupby('house')['points'].sum().sort_values(ascending=False)
```

```python
final.groupby('house')['points'].sum().reset_index().sort_values(by='points', ascending=False)
```

### 어떤 과목에서 주로 점수를 얻었는지 과목별 점수 총합을 구해봅시다.

```python
final.groupby('subject')['points'].sum().reset_index().sort_values(by='points', ascending=False)
```

### 각 기숙사별로 강한 과목이 있을까요? 피벗 테이블을 사용해서 각 기숙사가 어떤 과목에 강하고 약한지 파악해봅시다.

```python
pd.pivot_table(final, index='house', columns='subject', values='points')
```

```python
final.pivot_table(
    index='house',
    columns='subject',
    values='points',
    aggfunc='mean'
)
```

### 또 어떤 관점에서 데이터를 분석할 수 있을까요? 여러분만의 인사이트를 발견해봅시다!

```python
#'house'별 'year'별 점수
pd.pivot_table(final, index=['house','year'], columns='subject', values='points')
```

```python
#'house'별 'student_name'별 점수
pd.pivot_table(final, index=['house','student_name'], columns='subject', values='points')
```

### 과목별 난이도

```python
final.groupby('subject')['points'].std().sort_values(ascending=False)
#표준편차가 클수록 학생 실력 차이 혹은 과목/시험 난이도가 높음
```

```python
final.groupby('subject')['points'].mean().sort_values(ascending=False)
```

```python
final.groupby('subject')['points'].median().sort_values(ascending=False)
```

```python
final.groupby(['year', 'subject'])['points'].std().reset_index().sort_values(by='points', ascending=False)
```
