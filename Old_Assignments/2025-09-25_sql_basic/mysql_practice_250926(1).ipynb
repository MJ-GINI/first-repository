```python
# employees 테이블을 사용하여 각 부서(dept_id)별로 총 몇 명의 직원이 있는지, 
# 평균 연봉은 얼마인지 계산하고, 
# 평균 연봉이 7,500만원 이상인 부서의 정보만 필터링하여 조회하는 SQL 쿼리를 작성하세요.

mysql("""

SELECT dept_id, 
       COUNT(emp_id) AS '직원수',
       AVG(salary) AS '평균 연봉'
FROM employees
GROUP BY dept_id
         HAVING AVG(salary) >= 7500

""")
```

```python
#departments 테이블과 employees 테이블을 JOIN한 후, 
#부서 이름(dept_name )별로 직원 수를 세고, 
#그 결과를 직원 수가 많은 순서대로 정렬하여 보여주는 SQL 쿼리를 작성하세요.

mysql("""

SELECT d.dept_name AS '부서명',
       COUNT(e.emp_id) AS '직원수'
FROM departments AS d
JOIN employees AS e
     ON d.dept_id = e.dept_id
GROUP BY d.dept_name
ORDER BY 직원수 DESC;

""")
```

```python
import pandas as pd

# 'Departments' (부서) 데이터프레임 생성
departments_data = {
    'dept_id': ['D101', 'D102', 'D103'],
    'dept_name': ['개발팀', '기획팀', '마케팅팀']
}
departments = pd.DataFrame(departments_data)


# 'Employees' (직원) 데이터프레임 생성
employees_data = {
    'emp_id': [1, 2, 3, 4, 5, 6, 7],
    'emp_name': ['김개발', '박기획', '이마케', '최코딩', '정디자인', '한분석', '윤홍보'],
    'dept_id': ['D101', 'D102', 'D103', 'D101', 'D102', 'D101', 'D103'],
    'position': ['팀장', '팀장', '사원', '사원', '대리', '대리', '대리'],
    'salary': [9000, 8500, 5500, 6000, 7200, 7500, 6800],
    'hire_date': ['2020-01-15', '2019-08-01', '2022-03-10', '2023-05-20', '2021-11-22', '2021-02-18', '2022-09-30']
}
employees = pd.DataFrame(employees_data)

print("준비 완료")
```

```python
#employees 테이블에서 
#각 직원(emp_name)의 연봉(salary)과 속한 부서(dept_id)의 평균 연봉(AVG(salary) 기준)을 함께 조회하고, 
#연봉이 부서 평균 이상이면 ‘Above’, 미만이면 ‘Below’로 표시하세요. (CASE WHEN + 서브쿼리 활용)

mysql("""

SELECT 
    e.emp_name, e.salary, 
    e.dept_id,
    CASE 
        WHEN e.salary >= 
            (SELECT AVG(salary)
            FROM employees
            WHERE dept_id = e.dept_id)
        THEN 'Above'
        ELSE 'Below'
    END AS '부서 평균 연봉 기준'
FROM employees AS e

""")
```

```python
#employees 테이블에서 최근 입사한 상위 3명의 
#직원 이름(emp_name), 입사일(hire_date)과 부서 이름(dept_name)을 조회하세요. (JOIN 사용)

mysql("""

SELECT e.emp_name, e.hire_date, d.dept_name
FROM employees AS e
JOIN departments AS d
     ON e.dept_id = d.dept_id
ORDER BY e.hire_date DESC
LIMIT 3

""")
```

```python
#employees 테이블에서 연봉(salary)이 
#전체 직원 평균 이상인 직원의 이름과 연봉을 조회하세요. (서브쿼리 사용)

mysql("""

SELECT 
    e.emp_name, e.salary, 
    e.dept_id,
    CASE 
        WHEN e.salary >= 
            (SELECT AVG(salary)
             FROM employees
             WHERE dept_id = e.dept_id)
        THEN 'Above'
        ELSE 'Below'
    END AS '부서 평균 연봉 기준'
FROM employees AS e
ORDER BY '부서 평균 연봉 기준' DESC;

""")
```

```python
#employees 테이블에서 부서(dept_id)별로 서로 다른 직급(position)의 수(COUNT(DISTINCT position))를 조회하세요.

mysql("""

SELECT dept_id AS '부서', 
       COUNT(DISTINCT position) AS '직급수',
       GROUP_CONCAT(DISTINCT position) AS '직급 목록'
FROM employees
GROUP BY dept_id
ORDER BY dept_id

""")
```
