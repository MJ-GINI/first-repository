```python
#1번 문제
#Products 테이블에서 가격(price)이 500,000원 이상인 상품의 product_name과 price를 조회하는 SQL 쿼리를 작성해 보세요.

mysql("""

SELECT product_name, price
FROM products
WHERE price >= 500000

""")
```

```python
#2번 문제
#Customers 테이블에서 이름(customer_name)이 '김'씨 성을 가진 모든 고객의 정보를 조회하는 SQL 쿼리를 작성해 보세요.

mysql("""

SELECT *
FROM customers
WHERE customer_name LIKE "김%"

""")
```

```python
#3번 문제
#Products 테이블에서 카테고리(category)가 '가전'인 상품들을 가격(price)이 비싼 순서대로 정렬하여 모든 정보를 조회하는 SQL 쿼리를 작성해 보세요.

mysql("""

SELECT *
FROM products
WHERE category = "가전"
ORDER BY price DESC

""")
```

```python
#4번 문제
#Customers 테이블에 등록된 고객들이 거주하는 도시(city)의 종류를 중복 없이 모두 조회하는 SQL 쿼리를 작성해 보세요.

mysql("""

SELECT DISTINCT city
FROM customers;

""")
```

```python
#5번 문제
#5. `Products` 테이블에서 상품의 가격(`price`)에 따라 등급을 나누려고 합니다.
#    - 1,000,000원 이상이면 '고가'
#    - 500,000원 이상이면 '중가'
#    - 그 외에는 '보급형'
#    으로 표시하고 싶습니다. 상품의 `product_name`, `price`, 그리고 새로 만든 `price_grade`를 함께 조회하는 SQL 쿼리를 작성해 보세요.

mysql("""

SELECT product_name, price
        ,CASE WHEN price >= 1000000 THEN "고가"
                WHEN price >= 500000 THEN "중가"
                ELSE "보급형"
        END AS 'price_grade'
        
FROM products;

""")
```

```python
#6번 문제
#Products 테이블에서 상품명(product_name)에 '스마트'라는 단어가 포함되고, 
#가격(price)은 300,000원 미만인 상품을 찾아, 
#상품명과 가격을 조회하는 SQL 쿼리를 작성해 보세요.

mysql("""

SELECT product_name, price
FROM products
WHERE product_name LIKE "%스마트%"
AND price < 300000

""")
```

```python
#7번 문제
#Customers 테이블에서 고객의 거주 도시(city)가 '서울'이면 '수도권', 아니면 '지방'으로 구분하여 
#customer_name과 함께 area라는 별칭의 컬럼으로 보여주는 
#SQL 쿼리를 작성해 보세요. (CASE WHEN 사용을 권장합니다.)


mysql("""

SELECT customer_name,
        CASE WHEN city = "서울" THEN "수도권"
        ELSE "지방"
        END AS area
FROM customers

""")
```

```python
#8번 문제
#Customers 테이블에서 2024년 이후에 가입(signup_date)한 고객들을 대상으로, 
#거주 도시(city)가 '부산'이거나 '인천'인 고객의 이름(customer_name)과 도시(city)를 
#가입일이 최신인 순서대로 조회하는 SQL 쿼리를 작성해 보세요.

mysql("""

SELECT customer_name, city
FROM customers
WHERE signup_date >= '2024-01-01'
AND city IN ('부산', '인천')
ORDER BY signup_date DESC

""")
```

```python
#products 테이블에서 가격이 100,000원 이상 500,000원 이하인 상풍의 이름과 가격을 조회하는SQL 쿼리를 작성해보세요.

mysql("""

SELECT product_name, price
FROM products
WHERE price Between 100000 And 500000

""")
```

```python
#customers 테이블에서 최근에 가입한 고객 5명의 이름과 가입일을 조회하는 SQL 쿼리를 작성해보세요.

mysql("""

SELECT customer_name, signup_date
FROM customers
ORDER BY signup_date DESC
LIMIT 5

""")
```

```python
#products 테이블에서 가격이 100만원 이상이면 10% 할인, 50만원 이상이면 5% 할인, 나머지는 정가로 분류하고
#상품 이름, 가격과 할인율을 조회하는 SQL 쿼리를 작성해보세요.

mysql("""

SELECT product_name,
        price,
        CASE 
            WHEN price >= 1000000 THEN price * 0.1
            WHEN price >= 500000 THEN price * 0.05
            ELSE "no discount"
        END AS price_discount
FROM products

""")
```

```python
#customers 테이블에서 이름에 ‘우’가 들어가고 거주지가 서울인 고객의 이름과 가입일을 조회하는 SQL 쿼리를 작성해보세요.

mysql("""

SELECT customer_name, signup_date, city
FROM customers
WHERE customer_name LIKE '%우%'
OR city = '서울'
ORDER BY signup_date ASC

""")
```

```python
#products 테이블에서 가격에 따라 제품을 아래와 같이 분류하고, 
#상품 이름, 가격, 가격 등급(별칭 추가)을 조회한 후, 가격이 높은 순으로 정렬하는 SQL 쿼리를 작성해보세요.

#- 100만원 이상 : 고가
#- 50만원 이상 : 중가
#- 그 외 : 저가

mysql("""

SELECT product_name, price,
    CASE
        WHEN price >= 1000000 THEN '고가'
        WHEN price >= 500000 THEN '중가'
        ELSE '저가'
    END AS price_grade
FROM products
ORDER BY price DESC

""")
```
