```python
#Right join: 오른쪽, 모든 행, 왼쪽 일치 행 반환

df_a.merge(df_b, on = 'id', how = 'right')
```

```python
#Inner join: 공통키 반환 (*merge의 기본 설정)

df_a.merge(df_b, on = 'id', how = 'inner')
```

```python
#Outer join: 두 행을 반환, 불일치 NaN 값으로 채워짐

df_a.merge(df_b, on = 'id', how = 'outer')
```
